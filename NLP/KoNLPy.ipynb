##Hannanum

```python
from konlpy.tag import Hannanum
hannanum = Hannanum()
```

tagset - 품사의 종류 출력

```python
print(hannanum.tagset)
```

analyze - 각 토큰에 대해 다양한 형태학적 후보를 반환

```python
print(hannanum.analyze('내 이름은~ 예진이야~'))
```

morphs - 문장을 형태소 단위로 분석

```python
print(hannanum.morphs('내 이름은~ 예진이야~'))
```

```python
print(hannanum.morphs('내 이르믄~ 예지니야~'))
```

nouns - 명사 추출기

```python
print(hannanum.nouns('내 이름은~ 예진이야~'))
```

```python
print(hannanum.nouns('내 이르믄~ 예지니야~'))
```

pos - 형태소를 분석하고 결과를 태깅하여 출력

```python
print(hannanum.pos('내 이름은~ 예진이야~'))
```

##Kkma

```python
from konlpy.tag import Kkma
kkma = Kkma()
```

```python
print(kkma.tagset)
```

```python
print(kkma.morphs('내 이름은~ 예진이야~'))
```

```python
print(kkma.morphs('내 이르믄~ 예지니야~'))
```

```python
print(kkma.nouns('내 이름은~ 예진이야~'))
```

```python
print(kkma.pos('내 이름은~ 예진이야~'))
```

sentences - 문장 감지

```python
print(kkma.sentences('내 이름은~ 예진이야~ 치킨과 피자, 그리고 햄버거'))
```

```python
print(kkma.sentences('내 이름은~ 예진이야! 치킨과 피자, 그리고 햄버거'))
```

```python
print(kkma.sentences('내 이름은~ 예진이야. 치킨과 피자, 그리고 햄버거'))
```
